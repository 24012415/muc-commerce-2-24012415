## 3. 定义清洗规则

本项目采用以下规则：

| 问题 | 处理规则 | 理由 |
|---|---|---|
| 数值字段缺失 | 使用总体中位数填补 | 稳健且不将缺失误解为 0 |
| Phone / Mobile Phone | 统一为 Mobile Phone | 同一业务类别 |
| COD / Cash on Delivery | 统一为 Cash on Delivery | 同一业务类别 |
| CC / Credit Card | 统一为 Credit Card | 同一业务类别 |
| Mobile / Mobile Phone | 统一为 Mobile Phone | 同一业务类别 |
| 完全重复行 | 若存在则删除 | 完全相同的记录不增加信息 |
| 业务不合规值 | 记录并复核 | 本数据不应仅凭 IQR 直接删除 |

注意：不按 Churn 分组填补缺失值。

```python
NUMERIC_MISSING_COLS = [
    "Tenure",
    "WarehouseToHome",
    "HourSpendOnApp",
    "OrderAmountHikeFromlastYear",
    "CouponUsed",
    "OrderCount",
    "DaySinceLastOrder",
]

CATEGORY_MAPPINGS = {
    "PreferredLoginDevice": {
        "Phone": "Mobile Phone"
    },
    "PreferredPaymentMode": {
        "COD": "Cash on Delivery",
        "CC": "Credit Card"
    },
    "PreferedOrderCat": {
        "Mobile": "Mobile Phone"
    }
}
```

## 4. 编写可复用清洗函数

函数要求：

- 不直接修改传入的原始 DataFrame；
- 返回 cleaned_df 和 cleaning_log；
- 日志至少包含处理步骤、处理规则、处理前记录数、处理后记录数、影响记录数；
- 完成重复值处理、缺失值处理、类别标准化和必要的数据类型转换。

```python
def clean_ecommerce_data(data):
    """
    清洗电商用户行为数据。

    参数：
        data: 原始用户行为 DataFrame

    返回：
        cleaned_df: 清洗后的 DataFrame
        cleaning_log: 处理日志 DataFrame
    """
    # TODO：复制数据，避免覆盖原始数据
    cleaned_df = data.copy()
    logs = []
    # TODO：创建日志列表 logs
    # TODO：删除完全重复行，并记录日志
    before_dup = len(cleaned_df)
    cleaned_df = cleaned_df.drop_duplicates()
    after_dup = len(cleaned_df)
    affected_dup = before_dup - after_dup
    logs.append({
        "处理步骤": "删除完全重复行",
        "处理规则": "完全重复行直接删除，无信息增益",
        "处理前记录数": before_dup,
        "处理后记录数": after_dup,
        "影响记录数": affected_dup
    })
    # TODO：对 NUMERIC_MISSING_COLS 使用中位数填补，并记录每列影响数量
    for col in NUMERIC_MISSING_COLS:
        before_fill = len(cleaned_df)
        missing_cnt = cleaned_df[col].isna().sum()
        # 全局中位数填充，不按Churn分组
        med_val = cleaned_df[col].median()
        cleaned_df[col] = cleaned_df[col].fillna(med_val)
        logs.append({
            "处理步骤": f"中位数填充缺失值-{col}",
            "处理规则": "使用总体中位数填补，稳健不将缺失误判为0，不按Churn分组",
            "处理前记录数": before_fill,
            "处理后记录数": before_fill,
            "影响记录数": missing_cnt
        })
    # TODO：对 CATEGORY_MAPPINGS 完成类别标准化，并记录每条映射影响数量
    for col_name, map_dict in CATEGORY_MAPPINGS.items():
        before_map = len(cleaned_df)
        total_affected = 0
        for old_val, new_val in map_dict.items():
            cnt = (cleaned_df[col_name] == old_val).sum()
            total_affected += cnt
            cleaned_df[col_name] = cleaned_df[col_name].replace(old_val, new_val)
        logs.append({
            "处理步骤": f"类别标准化-{col_name}",
            "处理规则": "同义业务类别统一合并",
            "处理前记录数": before_map,
            "处理后记录数": before_map,
            "影响记录数": total_affected
        })
    # TODO：将 Churn 和 Complain 转为整数类型
     # Churn
    before_type = len(cleaned_df)
    churn_affect = 0
    if cleaned_df["Churn"].dtype != "int64":
        churn_affect = len(cleaned_df)
        cleaned_df["Churn"] = cleaned_df["Churn"].astype(int)
    logs.append({
        "处理步骤": "类型转换-Churn转int",
        "处理规则": "流失标识转为整数类型便于建模",
        "处理前记录数": before_type,
        "处理后记录数": before_type,
        "影响记录数": churn_affect
    })
    # Complain
    complain_affect = 0
    if cleaned_df["Complain"].dtype != "int64":
        complain_affect = len(cleaned_df)
        cleaned_df["Complain"] = cleaned_df["Complain"].astype(int)
    logs.append({
        "处理步骤": "类型转换-Complain转int",
        "处理规则": "投诉标识转为整数类型便于建模",
        "处理前记录数": before_type,
        "处理后记录数": before_type,
        "影响记录数": complain_affect
    })

    # 生成日志DataFrame
    cleaning_log = pd.DataFrame(logs)
    # TODO：返回 cleaned_df 与 cleaning_log
    return cleaned_df, cleaning_log
    pass
```

## 5. 数据转换与候选异常值检查

为便于第五天分析，请新增：

- TenureGroup：用户使用时长分层；
- IsMobileLogin：是否主要使用移动端登录；
- 候选异常值报告：WarehouseToHome、OrderCount、CashbackAmount。

候选异常值只记录，不在本项目中自动删除。

```python
def iqr_outlier_summary(series):
    """输出 IQR 候选异常值摘要。"""
    series = series.dropna()
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    return {
        "Q1": q1,
        "Q3": q3,
        "下限": lower,
        "上限": upper,
        "候选异常值数量": int(((series < lower) | (series > upper)).sum())
    }

# TODO：构建 tenure_bins、tenure_labels，并用 pd.cut 新建 TenureGroup
tenure_bins = [0, 6, 12, 24, np.inf]
tenure_labels = ["0-6个月", "6-12个月", "12-24个月", "24个月以上"]
cleaned_df["TenureGroup"] = pd.cut(
    cleaned_df["Tenure"],
    bins=tenure_bins,
    labels=tenure_labels,
    right=False
)
# TODO：新建 IsMobileLogin，移动端为 1，其他设备为 0
cleaned_df["IsMobileLogin"] = np.where(
    cleaned_df["PreferredLoginDevice"] == "Mobile Phone",
    1,
    0
)
# TODO：生成 outlier_report（每行对应一个待检查字段）
outlier_check_cols = ["WarehouseToHome", "OrderCount", "CashbackAmount"]
outlier_report = pd.DataFrame(
    [iqr_outlier_summary(cleaned_df[col]) for col in outlier_check_cols],
    index=outlier_check_cols
)
outlier_report = outlier_report.reset_index().rename(columns={"index": "字段名"})

print("===== 用户使用时长分层统计 =====")
print(cleaned_df["TenureGroup"].value_counts().sort_index())
print("\n===== 移动端登录分布 =====")
print(cleaned_df["IsMobileLogin"].value_counts())
print("\n===== 候选异常值报告 =====")
print(outlier_report)
```
